Issues with data  
- Year column contains non-year values  
- Convert year column from object to integer  
- Price column contains non-integer values
- Price has commas in its prices and is in object 
- Convert price column from object to integer  
- kms_driven column contains "kms" along with integer values  
- kms_driven column contains NaN values

```python
car=car[car['year'].str.isnumeric()]
```

```python
car['year']=car['year'].astype(int)
```

```python
car=car[car['Price']!='Ask For Price']
```

```python
car['Price']=car['Price'].str.replace(',','').astype(int)
```

```python
car['kms_driven']=car['kms_driven'].str.split().str.get(0).str.replace(',','')
```

```python
car['kms_driven']=car['kms_driven'].astype(int)
```

```python
car=car[~car['fuel_type'].isna()]
```

```python
car['name']=car['name'].str.split().str.slice(start=0,stop=3).str.join(' ')
```

# Model


```python

X=car[['name','company','year','kms_driven','fuel_type']]
y=car['Price']
```

```python
X
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.metrics import r2_score
```

```python

ohe=OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
column_trans=make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company','fuel_type']),
                                    remainder='passthrough')
```

```python
lr=LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
#checking r2 score
r2_score(y_test,y_pred)
```

Finding the model with a random state of TrainTestSplit where the model was found to give almost 0.92 as r2_score

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
pipe.predict(pd.DataFrame(columns=X_test.columns,data=np.array(['Maruti Suzuki Swift','Maruti',2019,100,'Petrol']).reshape(1,5)))
```

Best Model is found at a certain random state

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```
